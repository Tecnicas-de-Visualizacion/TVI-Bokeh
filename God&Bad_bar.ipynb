# Gráfico de barras

En este gráfico, los ejes no están etiquetados, por lo que no sabemos qué se está midiendo en el eje y. Además, los colores no se corresponden con los productos reales, lo que puede confundir al lector. Por otro lado, no se han agregado títulos o leyendas que describan el gráfico, lo que dificulta la comprensión del lector sobre lo que se está presentando. También es difícil comparar los valores de una ciudad a otra, ya que los intervalos de los ejes no están escalados adecuadamente.

```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource

# Datos de ejemplo
productos = ['Producto 1', 'Producto 2', 'Producto 3']
ciudades = ['Ciudad A', 'Ciudad B', 'Ciudad C']
ventas = [[10, 20, 15], [15, 10, 20], [20, 15, 10]]

# Crear una fuente de datos para el gráfico
source = ColumnDataSource(data=dict(productos=productos, ciudades=ciudades, ventas=sum(ventas,[])))

# Crear la figura del gráfico
p = figure(x_range=productos, y_range=(0, max(sum(ventas,[]))*1.1), plot_height=350, title="Ventas por producto y ciudad")

# Crear las barras del gráfico
p.vbar(x='productos', top='ventas', width=0.9, source=source, 
       line_color='white', fill_color='orange')

# Configurar la apariencia del gráfico
p.xgrid.grid_line_color = None
p.xaxis.axis_label = "Productos"
p.yaxis.axis_label = "Ventas"
p.xaxis.major_label_orientation = 1.2

# Mostrar el gráfico en el notebook
output_notebook()
show(p)
```

```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource
from bokeh.palettes import Category10
from bokeh.transform import factor_cmap

# Datos de ejemplo
frutas = ['Manzanas', 'Naranjas', 'Bananas', 'Peras']
cantidad = [20, 25, 30, 15]

# Crear una fuente de datos para el gráfico
source = ColumnDataSource(data=dict(frutas=frutas, cantidad=cantidad))

# Crear la figura del gráfico
p = figure(x_range=frutas, plot_height=350, title="Cantidad de frutas")

# Crear las barras del gráfico
p.vbar(x='frutas', top='cantidad', width=0.9, source=source, 
       line_color='white', fill_color=factor_cmap('frutas', palette=Category10[4], factors=frutas))

# Configurar la apariencia del gráfico
p.xgrid.grid_line_color = None
p.y_range.start = 0
p.yaxis.axis_label = "Cantidad"

# Mostrar el gráfico en el notebook
output_notebook()
show(p)
```
